```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def raiz(x):
  return 1/np.sqrt(x)
```

```python
def p(k, n):
  media = n * 0.5
  sigma2 = n * 0.25
  return 1/np.sqrt(2*np.pi*sigma2) * np.exp(-(k - media)**2 /(2*sigma2))
```

```python
def coin_game(n):

  k = 0

  for i in range(n):
    rand = random.random()

    if rand < 0.5:
      k += 1

  return k
```

```python
N = 10000
n_jogos = 50
estimador = n_jogos * 0.5

m2 = 0
m = 0
soma = []
erro = []

for i in range(N):
  k = coin_game(n_jogos)
  m += k
  m2 += k**2
  soma.append(m / (i+1))
  if i == 0:
    erro.append(0)
  else:
    av2 = (m2/(i+1))
    av = ((m/(i+1))**2)
    erro.append(np.sqrt((av2 - av) / i))

plt.plot(range(N), erro, label = "Erro Experimento")
plt.plot(range(N), raiz(range(N)), label = f"$1 / sqrt(x)$")
plt.plot(range(N), soma)
plt.legend()
#plt.axhline(y = estimador, color = 'black', ls='--')
plt.show()
```

```python
N = 10000
hist = np.zeros(N)

for i in range(N):
  k = coin_game(n_jogos)
  hist[i] += k

x = np.linspace(0,n_jogos, 100)

plt.hist(hist, bins=n_jogos+1, range=[-0.5, n_jogos + 0.5], color = 'green', rwidth=0.9, density=True)
plt.plot(x, p(x, n_jogos), color = 'red')
plt.show()
```
